```python
import pandas as pd
```

```python
df_stud = pd.DataFrame([
    [1, 'irina', '1a'],
    [2, 'olga', '4b'],
    [3, 'stas', 's5'],
    [5, 'tom', '3d']],
    columns = ['id', 'name', 'class'])
```

```python
df_stud
```

```python
df_mail = pd.DataFrame([
    [3, 'stas@mail.ru'],
    [2, 'olga@mmail.ru'],
    [2, 'olga@gmail.com'],
    [1, 'irina@inbox.com']],
    columns = ['id', 'e-mail'])
```

```python
df_mail
```

```python
df_new_stud = pd. DataFrame([[5, 'mark', '10a']], columns = ['id', 'name', 'class'])
```

```python
df_new_stud
```

добавление строк в датафрейм

```python
df_stud = df_stud.append(df_new_stud, ignore_index = True)
```

переименование столбца

```python
df_stud = df_stud.rename(columns = {'name':'Ferst name'})
```

```python
df_stud
```

объединение через конкатинацию

```python
pd.concat([df_stud, df_mail], axis = 1, join = 'outer', ignore_index = False)
```

функция merge

```python
pd.merge(df_stud, df_mail, on = 'id')
```

объединение по id

```python
pd.merge(df_stud, df_mail, on = 'id', how = 'inner')
```

Объединение по inner

```python
from numpy.core.numeric import outer
pd.merge(df_stud, df_mail, on = 'id', how = 'outer')
```

объединение через outer

```python
pd.merge(df_stud, df_mail, on = 'id', how = 'left')
```

объединение через left

```python
pd.merge(df_stud, df_mail, on = 'id', how = 'right')
```

объединение через right

```python
pd.merge(df_stud, df_mail, on = 'id', how = 'outer', indicator = 'True')
```

индикатор показывает расположение данных по таблицам

```python
df_stud.join(df_mail, lsuffix = '_left')
```
