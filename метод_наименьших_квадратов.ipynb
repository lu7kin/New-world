```python
import numpy as np
import matplotlib.pyplot as plt
```

Импортируем библиотеки

```python
N = 100   # число экспериментов
sigma = 3 # стандартное отклонение наблюдаемых значений
k = 0,5   # теоретическое значение параметра k (искомое)
b = 2     # теоретическое значение параметра b (искомое)
```

```python
x = np.array(range(N))                       # одномерный массив
f = np.array([k*z+b for z in range(N)])      # линия
y = f + np.random.normal(0, sigma, N)        # группа точек
```

  вычисляем коэффициенты k and b

```python
mx = x.sum()/N                  # сумма х разделенная на общее количество
my = y.sum()/N                  # сумма у, полученная экспериментальным путем/на общее число
a2 = np.dot(x.T, x)/N           # вычисляем второй момент
a11 = np.dot(x.T, y)/N          # вычисляем смешанный момент
```

```python
kk = (a11 - mx*my)/(a2 - mx**2)            # вычисляем к
bb = my - kk*mx                            # вычисляем в
ff = np.array([kk*z+bb for z in range(N)]) # линейный график
```

```python
plt.scatter(x, y, s=2, c='red')            # теоретическая прямая
plt.plot(f)                                # экспериментальная прямая
plt.plot(ff, c='red')                      # набор точек
plt.grid(True)
plt.show()
```

Теоретическая и экспериментальная прямые почти совпали, это значит, метод наименьших квадратов позволил вычислить наиболие близкие значения к и в.
